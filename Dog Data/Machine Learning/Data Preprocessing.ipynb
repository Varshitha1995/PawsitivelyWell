## Kaggle dataset : Data Preprocessing

```python
import pandas as pd
import numpy as np
from flask import Flask
```

```python
# Load the dataset
data = pd.read_csv('new_Data1.csv')
data.head(5)
```

```python
data.drop(['Country of Origin', 'Fur Color','Color of Eyes'], axis=1,inplace=True)
```

```python
data.head(10)
```

```python
#converting the range values to average 
def midpoint(value):
    try:
        val_min, val_max = map(int, value.split('-'))
        return (val_min + val_max) / 2
    except:
        return value
    

#Handling the string range values and 
def ignoreMonth(value):
    if isinstance(value, float):
        return value
    else:
        value = str(value)
        value = value.split('-')[0]
        return value
```

```python
data['Height'] = data['Height'].apply(midpoint)
data['Height'] = data['Height'].apply(ignoreMonth)
data.head(10)
```

```python
#converting the range values to average 
data['Age'] = data['Age'].apply(midpoint)

data['Age'] = data['Age'].apply(ignoreMonth)
data.head(10)
```

```python
# data = data.drop(['Color of Eyes'], axis=1)
# data.head(5)
```

```python

data['Weight'] = np.random.normal(loc=22, scale=10, size=len(data))
data['Weight'] = data['Weight'].round(1)
```

```python
data.isnull().sum()
```

```python
data.shape
data.head(5)
```

```python
# Saving the cleaned dataset
data.to_csv('Data3.csv', index=False)
```

## Web Scraped Data : Data Preprocessing

```python
# Load the dataset
data1 = pd.read_csv('new_Data2.csv')
data1.shape
data1.head(5)
```

```python
# Renaming columns 
data1 = data1.rename(columns={'Unnamed: 0': 'Breed'})
data1 = data1.rename(columns={'temperament': 'Character Traits'})
```

```python
# Removing unwanted columns
data1.drop(data1.columns[10:], axis=1, inplace=True)
data1.drop(['description', 'popularity'], axis=1,inplace=True)
data1.head(5)
```

```python
# Calculating Height and dropping some columns
data1 = data1.assign(Height=(data1['min_height'] + data1['max_height'])/2)
data1.drop(['min_height', 'max_height'], axis=1,inplace=True)
data1.head(5)
```

```python
# Calculating Weight and dropping some columns
data1 = data1.assign(Weight=(data1['min_weight'] + data1['max_weight'])/2)
data1.drop(['min_weight', 'max_weight'], axis=1,inplace=True)
data1.head(5)
```

```python
# Calculating Age and dropping some columns
data1 = data1.assign(Age=(data1['min_expectancy'] + data1['max_expectancy'])/2)
data1.drop(['min_expectancy', 'max_expectancy'], axis=1,inplace=True)
data1.head(5)
```

```python
data1[['Age', 'Weight']] = data1[['Age', 'Weight']].fillna(data1[['Age', 'Weight']].mean())
```

```python
# Saving the cleaned dataset
data1.to_csv('Data2.csv', index=False)
```
